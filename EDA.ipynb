## 2. Numerical Features

```python
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtype != 'O']
print('Number of numerical features: {}'.format(len(numerical_features)))
```

### Discrete Features

```python
discrete_features = [feature for feature in numerical_features if len(dataset[feature].unique())<25]
print('Number of discrete features: {}'.format(len(discrete_features)))
```

#### Observation-
No discrete variables exist, only continuous features exist.


```python
for feature in numerical_features:
    data = dataset.copy()
    data[feature].hist(bins=25)
    
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

#### Observation-
Our data is skewed. We'll convert the data into the log-normal form.

```python
for feature in numerical_features:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['price'] = np.log(data['price'])
        plt.scatter(data[feature], data['price'])
        plt.xlabel(feature)
        plt.ylabel('Price')
        plt.title(feature)
        plt.show()
```

## 4. Categorical variables

```python
categorical_features = [feature for feature in dataset.columns if dataset[feature].dtype == 'O']
categorical_features
```

```python
dataset[categorical_features].head()
```

### Cardinality value

```python
for feature in categorical_features:
    print(feature, ":", len(dataset[feature].unique()))
```

### Relationship between categorical and dependent variable

```python
for feature in categorical_features:
    data = dataset.copy()
    data.groupby(feature)['price'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Price')
    plt.title(feature)
    plt.show()
```

#### Observation- 
It is clear that bus_ter has only one category that is YES.

## 6. Outliers

```python
for feature in numerical_features:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column = feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

#### Observation-
1. There are a lot of outliers in age, room_num and the dependent variable 'price'
2. there are a few outliers in poor_props, n_hot_rooms, rainfall
